```python
prob1 = LpProblem("LC(1,2) Maximize", LpMaximize)
s1 = LpVariable("s1", lowBound = 0)
s2 = LpVariable("s2", lowBound = 0)
prob1 += 6*s1+7*s2
prob1 += 1*s1+1*s2<=20
prob1 += 2*s1+3*s2<=6
prob1
```

```python
status = prob1.solve()
LpStatus[status]
```

```python
value(prob1.objective)
```
